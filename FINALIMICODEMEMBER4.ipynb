```python
# Upload dataset
from google.colab import files
uploaded = files.upload()

# Load dataset
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

df = pd.read_csv("dataset_FINAL_READY.csv")

print("Original Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python

# --- Feature Engineering ---

# 1. Time to Threshold Failure
if "Impedance_Change_percent" in df.columns:
    df["Time_to_Threshold_Failure"] = 100 / (df["Impedance_Change_percent"] + 1e-6)

# 2. Log Impedance Ratio Normalized
if {"Final_Impedance_kOhm", "Initial_Impedance_kOhm"} <= set(df.columns):
    df["Log_Impedance_Ratio_Normalized"] = np.log1p(
        df["Final_Impedance_kOhm"] / (df["Initial_Impedance_kOhm"] + 1e-6)
    )

# 3. Surface Area Performance Score
if {"Surface_Area", "Conductivity_S_cm"} <= set(df.columns):
    df["Surface_Area_Performance_Score"] = df["Surface_Area"] * df["Conductivity_S_cm"]

# 4. Degradation Trend Slope Category
if "Impedance_Change_percent" in df.columns:
    df["Degradation_Trend_Slope_Category"] = pd.qcut(
        df["Impedance_Change_percent"], 3, labels=["Stable", "Moderate", "Rapid"]
    )

# 5. Normalized Degradation Percentile
if "Impedance_Change_percent" in df.columns:
    df["Normalized_Degradation_Percentile"] = df["Impedance_Change_percent"].rank(pct=True)

# 6. Anomaly Score (simple z-score)
if "Impedance_Change_percent" in df.columns:
    df["Anomaly_Score"] = (
        (df["Impedance_Change_percent"] - df["Impedance_Change_percent"].mean())
        / df["Impedance_Change_percent"].std()
    )

# 7. PCA Component
features = ["Charge_Injection_mC_cm2", "Frequency_Hz", "Surface_Area"]
available_features = [f for f in features if f in df.columns]

if available_features:
    scaled = StandardScaler().fit_transform(df[available_features])
    pca = PCA(n_components=1)
    df["Principal_Component_2"] = pca.fit_transform(scaled)
else:
    print("PCA skipped: required features not found in dataset.")
```

```python
# Save the updated DataFrame to a new CSV file
output_filename = "dataset_with_features.csv"
df.to_csv(output_filename, index=False)

print(f"File saved as {output_filename}")

from google.colab import files
files.download(output_filename)
```
